```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


amazon_data = pd.read_csv('AMZN.csv')
google_data = pd.read_csv('GOOGL.csv')
apple_data = pd.read_csv('AAPL.csv')
microsoft_data = pd.read_csv('MSFT.csv')
```

```python
# Combine data into a dictionary
stock_data = {
    'AMZN': amazon_data,
    'GOOG': google_data,
    'AAPL': apple_data,
    'MSFT': microsoft_data
}
```

```python
stock_data
```

```python
#  Perform Cleaning
for symbol, df in stock_data.items():
    df['Date'] = pd.to_datetime(df['Date'])
    df.dropna(inplace=True)
```

```python
# Change in Stock Price Over Time
plt.figure(figsize=(12, 6))
for symbol, df in stock_data.items():
    plt.plot(df['Date'], df['Adj Close'], label=symbol)

plt.title('Change in Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.legend()
plt.show()

```

```python
#  Visualize Change in Stock Volume Over Time
plt.figure(figsize=(12, 6))
for symbol, df in stock_data.items():
    plt.plot(df['Date'], df['Volume'], label=symbol)

plt.title('Change in Stock Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()

```

```python
#  Moving Average of Various Stocks
window = 30
plt.figure(figsize=(12, 6))
for symbol, df in stock_data.items():
    df['Moving Average'] = df['Adj Close'].rolling(window=window).mean()
    plt.plot(df['Date'], df['Adj Close'], label=symbol)
    plt.plot(df['Date'], df['Moving Average'], label=f'{symbol} {window}-Day Moving Avg')

plt.title('Stock Prices with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

```

```python
# Daily Return Average of a Stock
for symbol, df in stock_data.items():
    df['Daily Return'] = df['Adj Close'].pct_change()
    avg_daily_return = df['Daily Return'].mean()
    print(f'Average Daily Return for {symbol}: {avg_daily_return}')

```

```python
# Add a New Column 'Trend'
for symbol, df in stock_data.items():
    df['Trend'] = np.where(df['Daily Return'] >= 0, 'Up', 'Down')

```

```python
# Visualize Trend Frequency Through a Pie Chart
plt.figure(figsize=(6, 6))
for symbol, df in stock_data.items():
    df['Trend'].value_counts().plot(kind='pie', autopct='%1.1f%%', labels=None)
    plt.title(f'{symbol} Daily Return Trend')
    plt.legend(df['Trend'].unique(), loc='upper right')
    plt.show()

```

```python
#Correlation Between Daily Returns
# Create a DataFrame to store daily returns of all stocks
returns_df = pd.DataFrame({symbol: df['Daily Return'] for symbol, df in stock_data.items()})
correlation_matrix = returns_df.corr()

```

```python
# Visualize the correlation using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Daily Returns of Different Stocks')
plt.show()
```
